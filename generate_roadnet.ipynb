```python
################ test1 roadnet and flow
roadnet_path='./data/roadnet_test1.txt'
flow_path='./data/flow_test1.txt'
speedLimit=20
#######################################

import sys
import os
sys.path.append("agent/util")
from roadnet_util import RoadNet
from flow_util import FlowDataSetCreater,FlowDataSet
from graph_util import Graph,Node,Link

g=Graph()
delta=0.001
n_center=g.appendNode(120,30)
n_north=g.appendNode(120,30+delta*2)
n_east=g.appendNode(120+delta*2,30)
n_south=g.appendNode(120,30-delta*2)
n_west=g.appendNode(120-delta*2,30)
g.appendLink(n_center,n_north)
g.appendLink(n_center,n_east)
g.appendLink(n_center,n_south)
g.appendLink(n_center,n_west)
roadNet=RoadNet.createFromGraph(g,[n_center],speedLimit=speedLimit,onEarth=True)
txt=roadNet.getText()
print('----------roadnet-------------')
print(txt)
print('-----------------------')
roadNet.saveText(roadnet_path)

flowDSC = FlowDataSetCreater(roadNet.roadDataSet)
flowDSC.appendByInterIdList(0,10,5,[n_south,n_center,n_north])
flowDSC.appendByInterIdList(0,10,5,[n_north,n_center,n_south])

flowDSC.appendByInterIdList(30,40,5,[n_east,n_center,n_west])
flowDSC.appendByInterIdList(30,40,5,[n_west,n_center,n_east])

flowDSC.appendByInterIdList(60,70,5,[n_south,n_center,n_west])
flowDSC.appendByInterIdList(60,70,5,[n_north,n_center,n_east])

flowDSC.appendByInterIdList(90,100,5,[n_east,n_center,n_south])
flowDSC.appendByInterIdList(90,100,5,[n_west,n_center,n_north])

flowDS=flowDSC.compile()
txt = flowDS.getText()
print('----------flow-------------')
print(txt)
print('-----------------------')
flowDS.saveText(flow_path)

#########################################################
roadNet=RoadNet.createFromRoadNetFile(roadnet_path)
roadNet.exportAsJson(os.path.splitext(roadnet_path)[0]+".json")

flowDS=FlowDataSet.createFromFlowFile(flow_path)
flowDS.exportAsJson(os.path.splitext(flow_path)[0]+".json")
```
